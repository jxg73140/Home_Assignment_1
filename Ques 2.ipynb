```python
import numpy as np
import matplotlib.pyplot as plt

# 1. True values and model predictions
y_true = np.array([0, 1, 0])  
y_pred = np.array([0.2, 0.7, 0.1]) 

# 2. Define functions for MSE and CCE
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def categorical_cross_entropy(y_true, y_pred):
    return -np.sum(y_true * np.log(y_pred + 1e-15))  # Adding 1e-15 to avoid log(0)

# 3. Calculate loss values
mse_loss = mean_squared_error(y_true, y_pred)
cce_loss = categorical_cross_entropy(y_true, y_pred)

# Print initial loss values
print("Initial Loss Values:")
print(f"Mean Squared Error (MSE): {mse_loss}")
print(f"Categorical Cross-Entropy (CCE): {cce_loss}")

# 4. Modify predictions slightly
y_pred_modified = np.array([0.3, 0.6, 0.1])

# Calculate new loss values
mse_loss_modified = mean_squared_error(y_true, y_pred_modified)
cce_loss_modified = categorical_cross_entropy(y_true, y_pred_modified)

# Print modified loss values
print("\nModified Loss Values:")
print(f"Mean Squared Error (MSE) with modified predictions: {mse_loss_modified}")
print(f"Categorical Cross-Entropy (CCE) with modified predictions: {cce_loss_modified}")

# 5. Plot the loss values using Matplotlib
labels = ['MSE', 'CCE']
initial_losses = [mse_loss, cce_loss]
modified_losses = [mse_loss_modified, cce_loss_modified]

x = np.arange(len(labels))  # Label locations
width = 0.35  # Bar width

fig, ax = plt.subplots()
bar1 = ax.bar(x - width/2, initial_losses, width, label='Initial')
bar2 = ax.bar(x + width/2, modified_losses, width, label='Modified')

# Add labels and title
ax.set_xlabel('Loss Type')
ax.set_ylabel('Loss Value')
ax.set_title('Comparison of MSE and CCE Loss')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Show the plot
plt.show()
```
